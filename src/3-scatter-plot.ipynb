### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
import matplotlib.pyplot as plt
# Data for Scatter Plot
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]

# Scatter Plot
plt.figure(figsize=(8, 6))
plt.scatter(x_scatter, y_scatter, color='orange', label='Points')

# Annotate each point
for i, (x_val, y_val) in enumerate(zip(x_scatter, y_scatter)):
    plt.text(x_val, y_val, f'({x_val}, {y_val})', fontsize=10, ha='right')

# Customization
plt.title('Scatter Plot with Annotations', fontsize=16)
plt.xlabel('x values', fontsize=14)
plt.ylabel('y values', fontsize=14)
plt.legend(fontsize=12)

# Show the scatter plot
plt.show()
```
