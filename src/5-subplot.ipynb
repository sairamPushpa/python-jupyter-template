### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt
import numpy as np

# Data preparation
x = np.linspace(0, 10, 100)  # Generate 100 points between 0 and 10
y = x**2  # Compute y as the square of x
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'green', 'blue', 'purple']
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Line Plot in the first subplot
axs[0, 0].plot(x, y, label='y = x^2', color='blue')
axs[0, 0].set_title('Line Plot of y = x^2', fontsize=12)
axs[0, 0].set_xlabel('x values', fontsize=10)
axs[0, 0].set_ylabel('y values', fontsize=10)
axs[0, 0].grid(True, linestyle='--', alpha=0.7)
axs[0, 0].legend(fontsize=10)

# Bar Plot in the second subplot
axs[0, 1].bar(categories, values, color=colors)
axs[0, 1].set_title('Bar Plot of Categories and Values', fontsize=12)
axs[0, 1].set_xlabel('Categories', fontsize=10)
axs[0, 1].set_ylabel('Values', fontsize=10)

# Scatter Plot in the third subplot
axs[1, 0].scatter(x_scatter, y_scatter, color='orange', label='Points')
for i, (x_val, y_val) in enumerate(zip(x_scatter, y_scatter)):
    axs[1, 0].text(x_val, y_val, f'({x_val}, {y_val})', fontsize=8, ha='right')
axs[1, 0].set_title('Scatter Plot with Annotations', fontsize=12)
axs[1, 0].set_xlabel('x values', fontsize=10)
axs[1, 0].set_ylabel('y values', fontsize=10)
axs[1, 0].legend(fontsize=10)

# Pie Chart in the fourth subplot
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['gold', 'lightblue', 'lightgreen', 'pink'])
axs[1, 1].set_title('Pie Chart of Programming Languages', fontsize=12)

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
